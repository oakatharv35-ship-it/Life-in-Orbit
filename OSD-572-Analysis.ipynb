```python
def parse_housing_temp(val):
    try:
        return float(val)
    except Exception:
        if isinstance(val, str) and "Room Temperature" in val:
            return 22.0
        return np.nan

def parse_storage_temp(val):
    # Accepts "In-flight: RT; On-Earth: 4" or similar; averages numeric entries; treats RT as ~22 C
    if isinstance(val, str):
        parts = [p.strip() for p in val.replace("In-flight:", "").replace("On-Earth:", "").split(";")]
        nums = []
        for p in parts:
            if p.upper() == "RT":
                nums.append(22.0)
            else:
                try:
                    nums.append(float(p))
                except Exception:
                    pass
        if nums:
            return float(np.mean(nums))
    try:
        return float(val)
    except Exception:
        return np.nan
```

```python
import numpy as np
```

```python
# --- Clean time ---
if "Factor Value[Time]" in df.columns:
    df["Time_days"] = df["Factor Value[Time]"].astype(str).str.extract(r"([-+]?\d+(?:\.\d+)?)").astype(float)
else:
    df["Time_days"] = np.nan

# --- Clean temperatures ---
if "Parameter Value[Housing Temperature]" in df.columns:
    df["Housing_Temp_C"] = df["Parameter Value[Housing Temperature]"].apply(parse_housing_temp)
else:
    df["Housing_Temp_C"] = np.nan

if "Parameter Value[Sample Storage Temperature]" in df.columns:
    df["Storage_Temp_C"] = df["Parameter Value[Sample Storage Temperature]"].apply(parse_storage_temp)
else:
    df["Storage_Temp_C"] = np.nan

# --- Choose numeric columns of interest ---
read_depth_col = None
for cand in ["Parameter Value[Read Depth]_x", "Parameter Value[Read Depth]_y", "Parameter Value[Read Depth]"]:
    if cand in df.columns:
        read_depth_col = cand
        break

human_removed_col = None
for cand in ["Parameter Value[Human Reads Removed]_x", "Parameter Value[Human Reads Removed]_y", "Parameter Value[Human Reads Removed]"]:
    if cand in df.columns:
        human_removed_col = cand
        break
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Filter out rows where either column is "Not Applicable"
df_filtered = df[
    (df["Factor Value[Spaceflight]"] != "Not Applicable") &
    (df["Characteristics[Host Sex]"] != "Not Applicable")
]

# Build crosstab
ctab = pd.crosstab(
    df_filtered["Factor Value[Spaceflight]"],
    df_filtered["Characteristics[Host Sex]"]
)

# Plot grouped bars (side-by-side)
plt.figure()
ctab.plot(kind="bar", stacked=False)  # side-by-side instead of stacked
plt.title("Host Sex Distribution by Spaceflight")
plt.xlabel("SpaceFlight Status")
plt.ylabel("Total Samples Collected")
plt.legend(title="Host Sex")
plt.tight_layout()
plt.show()
```

```python

# --- Plot 3: Boxplot of Read Depth by Spaceflight ---
if read_depth_col and "Factor Value[Spaceflight]" in df.columns:
    plt.figure()
    groups = []
    labels = []
    for level in sorted(df["Factor Value[Spaceflight]"].dropna().unique().tolist()):
        groups.append(df.loc[df["Factor Value[Spaceflight]"] == level, read_depth_col].dropna().values)
        labels.append(str(level))
    if groups:
        plt.boxplot(groups, labels=labels, showfliers=False)
        plt.yscale("log")  # optional for wide ranges
        plt.title("Read Depth by Spaceflight Condition (" + read_depth_col + ")")
        plt.xlabel("Spaceflight")
        plt.ylabel("Read Depth (log scale)")
        plt.tight_layout()
    plt.show()
```

```python
# --- Plot 4: Histogram of Human Reads Removed (%) ---
if human_removed_col:
    plt.figure()
    data = df[human_removed_col].dropna().astype(float)
    plt.hist(data, bins=20)
    plt.title("Distribution of Human Reads Removed (%) (" + human_removed_col + ")")
    plt.xlabel("Human Reads Removed (%)")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
```

```python
# --- Plot 5: Scatter Time vs Read Depth ---
if read_depth_col and "Time_days" in df.columns:
    x = df["Time_days"]
    y = df[read_depth_col]
    mask = x.notna() & y.notna()
    if mask.sum() > 0:
        plt.figure()
        plt.scatter(x[mask], y[mask], s=12)
        plt.yscale("log")
        plt.title("Read Depth vs Days from Launch")
        plt.xlabel("Days from Launch (Time_days)")
        plt.ylabel("Read Depth (log scale)")
        plt.tight_layout()
        plt.show()
```

```python
# --- Plot 6: Correlation Heatmap (numeric) ---
num_df = df.select_dtypes(include=[float, int]).copy()
if not num_df.empty:
    corr = num_df.corr(numeric_only=True)
    plt.figure()
    im = plt.imshow(corr.values, aspect="auto")
    plt.colorbar(im, fraction=0.046, pad=0.04)
    plt.title("Correlation Heatmap (Numeric Columns)")
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90, fontsize=7)
    plt.yticks(range(len(corr.index)), corr.index, fontsize=7)
    plt.tight_layout()
    plt.show()
```

```python
# --- Plot 7: Mean Read Depth over Time (line) ---
if read_depth_col and "Time_days" in df.columns:
    temp = df[["Time_days", read_depth_col]].dropna().copy()
    if not temp.empty:
        temp = temp.groupby("Time_days", as_index=False)[read_depth_col].mean().sort_values("Time_days")
        plt.figure()
        plt.plot(temp["Time_days"], temp[read_depth_col])
        plt.title("Mean Read Depth over Time (Days from Launch)")
        plt.xlabel("Days from Launch (Time_days)")
        plt.ylabel("Mean Read Depth")
        plt.tight_layout()
        plt.show()
```
